## TextBlob:
  - Lexicon-based sentiment analyzer with a Bag of Words approach and some predefined rules to consider negation words. For a given text, it returns a polarity score ranging between -1.0 and +1.0, which is an average of the polarities associated with each word. Polarity scores equal 0.0 were considered neutral, while positive scores were considered positive and negative were considered negative.

```python
import nltk
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from sklearn.metrics import (
    accuracy_score,
    average_precision_score,
    classification_report, 
    f1_score
)


nltk.download([
    "vader_lexicon",
])
```

```python
def get_textblob_polarity(text):
    textblob = TextBlob(text)
    
    return textblob.sentiment.polarity
    
def get_textblob_sentiment(polarity):
    sentiment = 'neutral'
    if polarity > 0:
        sentiment = 'positive'
    elif polarity < 0:
        sentiment = 'negative'
    
    return sentiment


wdata['textblob_polarity'] = wdata['headline_snippet'].apply(get_textblob_polarity)
wdata['textblob_sentiment'] = wdata['textblob_polarity'].apply(get_textblob_sentiment)

print(wdata['textblob_sentiment'].value_counts())
wdata[["headline.main", "snippet", "textblob_sentiment",'textblob_polarity']].sample(5)
```

# Vader Sentiment classifier
- Short for "Valence Aware Dictionary and sEntiment Reasoner", VADER is also a lexicon and rule-based sentiment analysis tool based mostly on sentiments expressed in social media, but generally applicable to other domains. It was published by Hutto, & Gilbert in 2014, and associates different intensities (valences) for the sentiment of each word of its vocabulary, and has a built-in algorithm to change sentiment intensity based on punctuations, slang, emojis, and acronyms. The same threshold used for TextBlob's polarity score was used for VADER's compound score.

```python
sid = SentimentIntensityAnalyzer()
wdata['vader_polarity_scores'] = wdata['headline_snippet'].apply(sid.polarity_scores)

wdata[["headline.main", "snippet", "vader_polarity_scores"]]

def classify(vader_polarity_score):
    compound = vader_polarity_score['compound']
    sentiment = 'neutral'
    if compound > 0:
        sentiment = 'positive'
    elif compound < 0: 
        sentiment = 'negative'
    return sentiment

wdata['vader_sentiment'] = wdata['vader_polarity_scores'].apply(classify)

print(wdata["vader_sentiment"].value_counts())
```

# Word Cloud

```python
import string
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud


nltk.download([
    "punkt", 
    "stopwords", 
    "wordnet",
    'omw-1.4', 
])
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")

ignore_tokens=['covid','pandemic','coronavirus']
```

```python
def preprocess(text):
    """
    Apply the following Text Preprocessing steps to the text: 
    Lower key > Remove Punctuation > Tokenize >  
    Remove English Stopwords > Lemmatize 
    """
    text = str.lower(str(text))
    
    punctuations = string.punctuation + '’‘”“'
    text_p = "".join([char for char in text if char not in punctuations])
    
    words = word_tokenize(text_p)
    
    stop_words = stopwords.words('english') 
    filtered_words = [word for word in words if word not in stop_words]
    
    lemmatizer = WordNetLemmatizer()
    lemmatized = [lemmatizer.lemmatize(word) for word in filtered_words]
    
    return lemmatized

wdata["headline_snippet_prepr"] = wdata["headline_snippet"].apply(preprocess)
wdata[["headline.main", "snippet", "headline_snippet_prepr"]].sample(5)
```

```python
wordcloud_negative = WordCloud(
    width=1100, 
    height=900, 
    random_state=42,
    background_color='white', 
    colormap='rocket',
    stopwords=ignore_tokens
)

wordcloud_positive = WordCloud(
    width=1100, 
    height=900, 
    random_state=42,
    background_color='white', 
    colormap='mako',
    stopwords=ignore_tokens
)
```

```python
negative = wdata[(wdata['vader_sentiment']=='negative')
]['headline_snippet_prepr']
positive = wdata[(wdata['vader_sentiment']=='positive')
]['headline_snippet_prepr']
```

```python
print("Negative : \n")
print(negative)
print("=========================================================")
print("Positive : \n")
print(positive)
```

```python
wordcloud_negative.generate(' '.join(negative.apply(' '.join)))
wordcloud_positive.generate(' '.join(positive.apply(' '.join)))
fig, ax = plt.subplots(1,2, figsize=(16, 8))
ax[0].axis("off")
ax[1].axis("off")

fig.suptitle("Wordcloud")
ax[0].set_title("Negative")
ax[1].set_title("Positive")

ax[0].imshow(wordcloud_negative)
ax[1].imshow(wordcloud_positive)
plt.tight_layout()
```
